```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.preprocessing import OneHotEncoder
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Don't touch
#! pop 

X_train = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv').set_index(['Id'])
y_train = X_train.pop('SalePrice')
y_train
```

```python
X_train.select_dtypes(include=object)
```

```python
X_train.select_dtypes(exclude=object)
X_num_train =  X_train.select_dtypes(exclude=object).columns.tolist()
X_num_train
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# std_sc = StandardScaler()
std_sc = MinMaxScaler()
X_train[X_num_train] = std_sc.fit_transform(X_train[X_num_train])
```

```python
X_ctg_train = X_train.select_dtypes(include=object).columns.tolist()
#X_ctg_train
X_train[X_ctg_train] = X_train[X_ctg_train].astype("category")
X_train.info()
```

```python
to_drop = ['Alley', 'PoolQC', 'Fence', 'MiscFeature']
X_train.drop(to_drop, inplace=True, axis=1)
X_train.shape
```

```python
model = XGBRegressor(tree_method="gpu_hist", enable_categorical=True, use_label_encoder=False)
model.fit(X_train, y_train)
```

```python
X_test = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv').set_index('Id')
X_test.drop(to_drop, inplace=True, axis=1)

X_test
```

```python
X_num_test =  X_test.select_dtypes(exclude=object).columns.tolist()
X_num_test

X_test[X_num_test] = std_sc.fit_transform(X_test[X_num_test])
```

```python
X_ctg_test = X_test.select_dtypes(include=object).columns.tolist()
#X_ctg_train
X_test[X_ctg_test] = X_test[X_ctg_test].astype("category")
X_test.info()
```

```python
X_test
```

```python
y_test_pred = model.predict(X_test)
y_test_pred
```

```python
df_res = pd.DataFrame(y_test_pred, index=X_test.index, columns=['SalePrice'])
df_res
```
